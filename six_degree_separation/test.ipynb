```python
import numpy as np
import network_model
import matplotlib.pyplot as plt

def PlotHist(L, Z, p):
    net = network_model.network(L, Z, p)
    net.RandomAdd()
    net.Display()
    distance_list = net.FindAllPathLengths()
    fig, ax = plt.subplots()
    ax.hist(distance_list, bins=25)
```

```python
x = [10**(i/10) for i in range(-30, 1)]
y = []
net = network_model.network(50, 2, 0)
net.RandomAdd()
_, regular_net = net.FindAveragePathLength()
for i in range(len(x)):
    L = 50
    Z = 2
    p = x[i]
    net = network_model.network(L, Z, p)
    net.RandomAdd()
    _, avg_path = net.FindAveragePathLength()
    avg_path = avg_path / regular_net
    print(p, avg_path)

    y.append(avg_path)
```

```python
plt.semilogx(x, y, marker="o")
```
